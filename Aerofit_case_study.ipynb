Based on martial status

```python
df.head()
```

```python
plt.figure(figsize=(8, 8))
sns.countplot(data=df,x='Product',hue='Gender',palette='BuGn_r')
plt.title(" Product purchased by customer based on Gender")

plt.show()
```

Among 3 products , KP281 and KP481 have almost same number of both male and female customers. But KP781 have purchased more by male.

```python
product_gender = pd.crosstab(df['Product'], df['Gender'])
product_gender.head()
```

```python
conditional_probability_product_gender = product_gender.div(product_gender.sum(axis=1), axis=0)
conditional_probability_product_gender
```

*   For Product KP281 , we can see that there is 50% probability that is purchased by either male or female.

*  For Product KP481 , we can see that there is almost 50% probability that is purchased by either male or female.

*  For Product KP781, we can see that there is higher chance that it was purchased by male (80%) .


```python
marginal_probability_product_gender = product_gender.div(product_gender.sum(axis=0), axis=1)
marginal_probability_product_gender
```

*  Among Female customers , nearly 52% purchased product KP281 and 38% of them chooses KP481 and least percent of female(9%) has purchased product KP781.
*  Among male customers , there is fair distributions among the products purchased, 38% purchased product KP281 and nearly 29% of them chooses KP481 and 31% has purchased product KP781.
